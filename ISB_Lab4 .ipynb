В предлагаемом для лабораторной образе использован относительно свежий алгоритм yescrypt.

Не исключено, что вам придется разобраться в параметрах его работы, об этом можно почитать [тут](https://unix.stackexchange.com/questions/690679/what-does-j9t-mean-in-yescrypt-from-etc-shadow).

При помощи [модуля](https://pypi.org/project/pyescrypt/), реализующий данный алгоритм хеширования, можно решить предлагаемую в лабораторной задачу.Если данный модуль не работает на вашей виндовой машине, но выполняйте лабораторную в Google Colab. Небольшой пример использования этого модуля для экономии вашего времени:

```python
from pyescrypt import Yescrypt, Mode

# Библиотека дает возможность получать результат хеширования в 3х форматах: 
# JSON, RAW и MCF. 
# Посмотрите на результат работы бибилиотеки в каждом из режимов.
# Как только вы обнаружите пароли на .vdi диске,
# будет очевидно, какой формат выбрать.
hasher = Yescrypt(mode=Mode.MCF)

# Если вы внимательно прочитаете теоретическую справку 
# о формате хранения паролей в Unix-системах, вам также станет вполне ясно,
# какие данные передать в качестве settings
hash = hasher.digest(password=b'some password', settings=b'some settings')
```
